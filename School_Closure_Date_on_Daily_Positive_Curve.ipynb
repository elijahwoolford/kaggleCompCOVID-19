```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from numpy import exp, loadtxt, pi, sqrt
```

```python
school_clos_df = pd.read_excel(r'data/coronavirus-school-closures-data.xlsx', header = 1)
```

```python
response_current = requests.get("https://covidtracking.com/api/states")
response_daily = requests.get("https://covidtracking.com/api/states/daily")
dfs = []
for data in enumerate(response_daily.json()):
    dfs.append(pd.DataFrame(data[1], index=[data[0]]))
df_final = pd.concat(dfs)
df_final['date_str']=df_final['date'].apply(lambda x: str(x))
```

```python
state_list = df_final['state'].unique()
```

```python
fig, axs = plt.subplots(12, 5, figsize = (25, 100))
axs = axs.ravel()
#fig.suptitle('States Daily Positive with School Closure Date', fontsize=16)
for i in range(len(state_list)):
    st_df = df_final[df_final["state"] == state_list[i]].sort_values(by="date", ascending=True)
    # school closure start date
    x = str(school_clos_df[school_clos_df['State Abbreviation'] == state_list[i]]['State Closure Start Date'].values[0])[0:10].replace('-', '')
    if x != 'NaT' and len(x):
        y = st_df[st_df['date_str'] == x]['positive'].values[0]
        axs[i].plot(st_df["date_str"], st_df["positive"])
        axs[i].scatter(x, y, color = 'red', s = 60)
    else:
        axs[i].plot(st_df["date_str"], st_df["positive"])
    axs[i].set_title(state_list[i], fontsize = 20)
    axs[i].set_xlabel('Date')
    #axs[i].set_xticklabels(rotation=90, labels = st_df["date_str"].values)
```
