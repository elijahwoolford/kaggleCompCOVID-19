# SEIR Model 
### reference website:  
### 1. datahubbs.com/social-distancing-to-slow-the-coronavirus/ # pythoon code for the SEIR model
### 2. https://www.thelancet.com/journals/langlo/article/PIIS2214-109X(20)30074-7/fulltext # for incubation period
### 3. https://arxiv.org/pdf/2002.06563.pdf #  for getting gamma value
### 4. https://www.healthline.com/health/r-nought-reproduction-number for explaining R0

```python
def base_seir_model(init_vals, params, t):
    S_0, E_0, I_0, R_0 = init_vals
    S, E, I, R = [S_0], [E_0], [I_0], [R_0]
    alpha, beta, gamma = params
    dt = t[1] - t[0]
    for _ in t[1:]:
        next_S = S[-1] - (beta*S[-1]*I[-1])*dt
        next_E = E[-1] + (beta*S[-1]*I[-1] - alpha*E[-1])*dt
        next_I = I[-1] + (alpha*E[-1] - gamma*I[-1])*dt
        next_R = R[-1] + (gamma*I[-1])*dt
        S.append(next_S)
        E.append(next_E)
        I.append(next_I)
        R.append(next_R)
    return np.stack([S, E, I, R]).T
```

```python
# get the population data
test = pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population')
population_df = test[0]
new_header = population_df.iloc[0] #grab the first row for the header
population_df = population_df[1:] #take the data less the header row
population_df.columns = new_header #set the header row as the df header

population_df = population_df[['State', 'Population estimate, July 1, 2019[2]']]
population_df.rename(columns = {'Population estimate, July 1, 2019[2]': 'Population'}, inplace = True)
```

```python
def State_level_projection(population_df, positive_df, state= "New York", state_abr = 'NY', R0 = 6, 
                           incubation_days = 5, infectious_days = 2) -> np.ndarray:
    
    N = int(population_df[population_df['State'] == state]['Population'].values[0]) # the population

    state_df = positive_df[positive_df['state'] == state_abr].sort_values('date') # real data

    I_0 = (state_df['positive'].values[0])/N # Infected,
    S_0 = 1 - I_0/N  # Susceptible  
    R_0 = 0 # Recovered
    E_0 = 0 # Exposed

    init_vals = S_0, E_0, I_0, R_0


    # Define parameters
    t_max = len(state_df)
    dt = 1
    t = np.linspace(0, t_max, int(t_max/dt) + 1)
#     incubation_days = 5 # incubation period
#     infectious_days = 2 # mean infectious period
#     R0 = 6 # the average number of people who will catch a disease from one contagious person.


    alpha = 1/incubation_days # the inverse of the incubation period 
    gamma = 1/infectious_days # the inverse of mean infectious period 
    beta = R0*gamma # average contact rate in the population ( = R0 * gamma = 8 * 0.5)

    params = alpha, beta, gamma
    # Run simulation
    results = base_seir_model(init_vals, params, t)
    
    return results
    
```

```python
def plot_projection(population_df, positive_df, school_clos_df, state= "New York", state_abr = 'NY',
                    incubation_days = 5, infectious_days = 2, R0_range = [4, 5]):
    
    df_1 = State_level_projection(population_df, df_final, state= state, state_abr = state_abr ,
                                R0 = R0_range[0], incubation_days = incubation_days, 
                                infectious_days = infectious_days)
    df_2 = State_level_projection(population_df, df_final, state= state, state_abr = state_abr ,
                                R0 = R0_range[1], incubation_days = incubation_days, 
                                infectious_days = infectious_days)
    state_df = positive_df[positive_df['state'] == state_abr].sort_values('date')
    
    N = int(population_df[population_df['State'] == state]['Population'].values[0])
    
    # get the school close date
    x = str(school_clos_df[school_clos_df['State Abbreviation'] == state_abr]['State Closure Start Date'].values[0])[0:10].replace('-', '')
    school_close_date = datetime.strptime(x, '%Y%m%d')
    report_start_date = datetime.strptime(str(state_df['date'].values[0]), '%Y%m%d')
    delta_date = school_close_date - report_start_date
    
    if x != 'NaT' and len(x):
        y = state_df[state_df['date_str'] == x]['positive'].values[0]
    
    #plt.plot(infected_projection_R0_6, color = 'b')
    plt.plot(df_1[1:, 2]*N, color = 'g')
    plt.plot(df_2[1:, 2]*N, color = 'b')
    plt.plot(t[1:], state_df['positive'], color = 'r')
    plt.scatter(delta_date.days, y, color = 'black', s = 60)
    legend_list = []
    plt.legend(['Project_R0='+str(R0_range[0]), 'Project_R0='+str(R0_range[1]),'Real Positive', 'School Close'])
    plt.xlabel('Days to the first day of reporting positive cases')
    plt.ylabel('Positive Cases')
    plt.title(state)
```
